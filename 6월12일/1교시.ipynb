```python
#1차 배열인데 간격을 정해서 넣어줘.

import numpy as np

# 기본 사용법
arr1 = np.linspace(0, 1, num=5)
print(arr1)
# 출력: [0.   0.25 0.5  0.75 1.  ]

# 개수와 간격 확인
arr2, step = np.linspace(0, 1, num=5, retstep=True)
print(arr2)
# 출력: [0.   0.25 0.5  0.75 1.  ]
print(step)
# 출력: 0.25

# 종료 값 포함하지 않음
arr3 = np.linspace(0, 10, num=5, endpoint=False)
print(arr3)
# 출력: [0. 2. 4. 6. 8.]

# 데이터 유형 지정
arr4 = np.linspace(1, 5, num=5, dtype=np.int)
print(arr4)
# 출력: [1 2 3 4 5]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x, y 데이터 생성
x = np.arange(0, 10, 0.1) #0부터 10까지 0.1씩 커지는. 배열. 그냥 0, 10 하면 10개만 찍어.
y = np.sin(x) #다 배열

# 선 그래프 그리기
plt.plot(x, y, marker='o') #플롯 함수는 선으로 이어줘. 마커는 점을 찍어줘.

# 그래프 타이틀과 축 라벨 설정
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('y')

# 그래프 출력
plt.show()

print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 색상 변경 예제
x = np.arange(0, 10, 0.1)
y1 = np.sin(x) #정의역은 같지만 공역이 다르니 2개로 나눠서.
y2 = np.cos(x)

plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='#FF5733')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, linestyle='dashed', label='Sine')
plt.plot(x, y2, linestyle='dashdot', label='Cosine')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', fontsize=12, shadow=True, title='Legend')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 스타일 리스트
styles = ['bmh', 'classic', 'dark_background', 'fivethirtyeight', 'ggplot', 'grayscale', 'Solarize_Light2', 'tableau-colorblind10']

# 스타일 적용 예제
for style in styles:
    plt.style.use(style)
    x = np.arange(0, 10, 0.1)
    y1 = np.sin(x)
    y2 = np.cos(x)

    plt.plot(x, y1, label='Sine')
    plt.plot(x, y2, label='Cosine')
    plt.title('Sine and Cosine Waves')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 2x2 플롯 예제
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 2x2 플롯 생성
fig, axs = plt.subplots(2, 2, figsize=(8, 8)) #서브플롯츠! 다 S 붙은거

#이건 일단 다 잘라놓고 같은 크기에서 그냥 집어넣는건데

# 첫 번째 서브플롯
axs[0, 0].plot(x, y1) #그래프 개수를 2차 형렬로 봐. 0,0이면 왼쪽위 싸인 그래프가 되겠지?
axs[0, 0].set_title('Sine')

# 두 번째 서브플롯
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine')

# 세 번째 서브플롯
axs[1, 0].plot(x, y1 + y2)
axs[1, 0].set_title('Sine + Cosine')

# 네 번째 서브플롯
axs[1, 1].plot(x, y1 - y2)
axs[1, 1].set_title('Sine - Cosine')

plt.show()

```

```python
import matplotlib.pyplot as plt

# 그리드 설정
fig = plt.figure()

#같은 캔버스에 각 그림을 각각 다르게 잘라. 매번 잘라.

#이건 s 안붙은 subplot
ax1 = plt.subplot(1, 3, 1) #1행 3열로 잘라서 첫번째에
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3]) #그중 첫번째 그래프를 이렇게 그리겠다~

ax2 = plt.subplot(2, 3, 2) #2행 3열로 잘라서 두번째에 그리겠다
ax2.plot([1, 2, 3, 4], [1, 4, 2, 3])
#매번 그릴때마다 이 창을 몇조각 해서 어떻게 넣을지를 정해서 한다
#ㅁㅁㅁ  -=> 123 순서니까. 2행 3열로 자른 이 행렬에서 2번째 칸에 넣는거.
#ㅁㅁㅁ  -=> 456


ax3 = plt.subplot(2, 3, 5)
ax3.plot([1, 2, 3, 4], [1, 4, 2, 3])

ax4 = plt.subplot(1, 3, 3)
ax4.plot([1, 2, 3, 4], [1, 4, 2, 3])

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 그리드 설정
fig = plt.figure()

# 첫 번째 서브플롯 (3x3 그리드의 (0, 0) 위치, 열 3칸 차지)
ax1 = plt.subplot2grid((3, 3), (0, 0), 1, 3)
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3])

# 두 번째 서브플롯 (3x3 그리드의 (1, 0) 위치, 열 2칸 차지)
ax2 = plt.subplot2grid((3, 3), (1, 0), 1,2)
ax2.plot([1, 2, 3, 4], [1, 4, 2, 3])

# 세 번째 서브플롯 (3x3 그리드의 (1, 2) 위치, 행 2 칸 차지)
ax3 = plt.subplot2grid((3, 3), (1, 2), 2, 1)
ax3.plot([1, 2, 3, 4], [1, 4, 2, 3])

# 네 번째 서브플롯 (3x3 그리드의 (2, 0) 위치, 열 2 칸 차지)
ax4 = plt.subplot2grid((3, 3), (2, 0), 1, 2)
ax4.plot([1, 2, 3, 4], [1, 4, 2, 3])

# 그래프 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(2010, 2021)  # 연도 데이터
gdp_a = np.array([100, 120, 150, 160, 180, 200, 220, 240, 260, 280, 300])  # A 나라의 연도별 GDP
gdp_b = np.array([80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180])  # B 나라의 연도별 GDP
gdp_c = np.array([200, 220, 240, 250, 260, 270, 280, 290, 300, 310, 320])  # C 나라의 연도별 GDP

bar_width = 0.25
index = np.arange(len(years))
print(index)
plt.style.use('default')
plt.bar(index-0.25, gdp_a, 0.25, label='ContryA') #바 넓이 변경가능(제일끝변수)
plt.bar(index, gdp_b, 0.25, label='ContryB') #인덱스에 플러스 마이너스로 인덱스 위치 조절로 그래프 위치 변경가능
plt.bar(index+bar_width, gdp_c, 0.25, label='ContryC')

plt.xticks(index+0.25, years) #인덱스 위치에 years 값들을 넣어라
#이러면 x축이 0,1,2,3,4 이런게 년도로 바뀐다.
#근데 위치가 좀 안좋네 

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 그리드 설정

#fig = plt.figure(figsize = (30,20)) 그래프 사이즈 지정가능

years = np.arange(2010, 2021)
gdp = np.array([100, 120, 150, 160, 180, 200, 220, 240, 260, 280, 300])
sales = np.array([50, 70, 30, 45, 60, 80, 70, 90, 110, 100, 120])
prices = np.array([10, 12, 15, 16, 18, 20, 22, 24, 26, 28, 30])


# 첫 번째 서브플롯 (3x3 그리드의 (0, 0) 위치, 열 3칸 차지)



ax1 = plt.subplot2grid((3, 3), (0, 0), 1, 3)
ax1.plot(years, gdp, marker = 'o', color='blue')
ax1.set_title('gdp')
plt.xlabel('years')
plt.ylabel('gdp')


# 두 번째 서브플롯 (3x3 그리드의 (1, 0) 위치, 열 2칸 차지)
ax2 = plt.subplot2grid((3, 3), (1, 0), 1,2)
ax2.bar(years, sales, color='green')
ax2.set_title('sales')
plt.xlabel('years')
plt.ylabel('sales')

# 세 번째 서브플롯 (3x3 그리드의 (1, 2) 위치, 행 2 칸 차지)
ax3 = plt.subplot2grid((3, 3), (1, 2), 2, 1)
ax3.scatter(years, prices, color='r')
ax3.set_title('sales')
plt.xlabel('years')
plt.ylabel('prices')

# 네 번째 서브플롯 (3x3 그리드의 (2, 0) 위치, 열 2 칸 차지)
ax4 = plt.subplot2grid((3, 3), (2, 0), 1, 2)
ax4.plot(years, sales, marker = 'o', color='blue')
ax4.set_title('sales')
plt.xlabel('years')
plt.ylabel('sales')


# 그래프 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
