```python
number = 23
guess = int(input('숫자입력 : '))


if guess == number:
  print ('축하염')
  print('근데 상은 읎다')

elif guess < number:   #가독성이 좋아졌당
  print('좀 더 큰 수다')
else:
  print('좀 더 작은 수다')

print('끝')
```
