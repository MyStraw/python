```python
#whille도 else 쓴다. 굳이 왜??? while에서 break를 한다면, else가 실행이 안된다. 만약 else없이 하면 그냥 무조건 실행되니까.
number = 23
running = True

while running:
  guess = int(input('숫자입력 : '))


  if guess == number:
    print ('축하염')
    print('근데 상은 읎다')
    running = False
    break

  elif guess < number:   #가독성이 좋아졌당
    print('좀 더 큰 수다')
  else:
    print('좀 더 작은 수다')

else: #굳이 하고싶어? 있어야 하나? break 할때를 생각해봐. break를 하면 여기 실행안됨.
  print("while 반복문 끝")

print('끝')


# If there is an else clause for a while loop, it is always executed
# unless you break out of the loop with a break statement
```

```python
#숫자 입력된게 5개 이상이면 5개 이상입니다.

count = 0
total = 0

print("-1 입력시 끝")
num = int(input("양의정수 입력하셈"))
min_val = num
max_val = num #max 라는 이름으로 안하는건 max는 함수가 있다. 덮어써버리면 이 함수를 못쓴다.

while num != -1:
  count += 1
  total += num
  if num < min_val:
    min_val = num
  if num > max_val:
    max_val = num
  num = int(input("양의정수 입력하셈"))

if count > 0 and count <5 :
  #if count>=5:
     #print(f'5개 이상으로 입력했구먼 {count}개를 입력했어')
  print('최소값', min_val)
  print('최대값', max_val)
  print('평균', total/count)
elif count >= 5:
  print(f'5개 이상으로 입력했구먼 {count}개를 입력했어')
  print('최소값', min_val)
  print('최대값', max_val)
  print('평균', total/count)
else:
  print('-1만 입력할거면 이거 왜하나?')
```

```python
sentence = input('Enter a sentence : ')
word1 = input('Enter word to replace : ')
word2 = input('Enter replacement word : ')

replaced = sentence.replace(word1, word2)
print(replaced)

#근데 이러면 단어를 pineapple 에서 apple을 바꾸면 안되는데 이거까지 바꾼다. apple만
```

```python
sentence = input('Enter a sentence : ')
word1 = input('Enter word to replace : ')
word2 = input('Enter replacement word : ')

words = sentence.split() #리스트로 만들어주고.
replaced = []

for w in words:
    if w == word1 :
        w = word2
    replaced.append(w)

replaced = ' '.join(replaced)

print(replaced)
```

```python
shoplist = ['apple', 'mango', 'carrot', 'banana']
mylist = shoplist
del mylist[0]
print(shoplist)
print(mylist)
#이건 메모리에 ['apple', 'mango', 'carrot', 'banana'] 이거 하나 만들어지고
#mylist랑 shoplist의 주소가 같다.
print(id(shoplist))
print(id(mylist)) #값이 들어있는 주소를 보니까 주소가 같다.
```

```python

#하지만 이건 메모리에 새로운 값을 하나 더 만든다. 그리고 각각을 가리킨다.
mylist = shoplist[:] #모두 슬라이스 해서 값을 가져온다음 만드는거라 새로 만드는 효과가 있다
del mylist[0]
print(shoplist)
print(mylist)

print(id(shoplist))
print(id(mylist)) #이건 주소가 다르네
```

```python
mylist = shoplist.copy() #이것 또한 새로 만드는거네 복사해서.
del mylist[0]
print(shoplist)
print(mylist)

print(id(shoplist))
print(id(mylist)) #이건 주소가 다르네
```
