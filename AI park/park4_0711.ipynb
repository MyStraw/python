```python
# 124p, 26, 30번 문제

def main():
    full = input("문자열 입력하셈")
    sub = input("substring 검색하고픈거 입력하셈")
    print(f"[문자열 : {full}], [섭문자 : {sub}], [갯수 : {sub_count(full, sub)}]")


def sub_count(full, sub):
    count = 0
    index = 0
    while index < len(full):
        if full[index:index + len(sub)] == sub:
            count += 1
            index += len(sub)
        else:
            index += 1
    return count

main()

```

```python
def displayWithListComprehension(file):
    infile = open("AIP_05_Data Processing and File Access_data/"+file, 'r')
    listPres = [line.rstrip() for line in infile]
    infile.close()
    print(listPres)
    
displayWithListComprehension('FirstPresidents.txt')
```

```python
def displayWithReadline(file):
    infile = open("AIP_05_Data Processing and File Access_data/"+file, 'r')
    line = infile.readline()
    while line != "":
        print(line.rstrip())
        line = infile.readline()
    infile.close()
    
displayWithReadline('FirstPresidents.txt')
    
```

```python
def main():
    L = ["George Washington", "John Adams", "Thomas Jefferson"]
    outfile = open("FirstPresidents2.txt", 'w')
    createWithWrite(L, outfile)
    outfile = open("FirstPresidents3.txt", 'w')
    createWithWriteLines(L, outfile)


def createWithWrite(L, outfile):
    for i in range(len(L)):  # 그냥 L 해도...
        outfile.write(L[i]+"\n")
    outfile.close()


def createWithWriteLines(L, outfile):
    for i in range(len(L)):
        L[i] = L[i] + "\n"
    outfile.writelines(L)
    outfile.close()


main()
```

```python
#위에 파일 만들때 대통령과 부대통령도 했던 사람 뽑는걸 교집합(intersection으로)
def main():
    vicePresList = createListFromFile("VPres.txt")
    presList = createListFromFile("USPres.txt")
    createNewFile(vicePresList, presList, "Both.txt")

def createListFromFile(fileName):
    infile = open("AIP_05_Data Processing and File Access_data/"+fileName, 'r')
    #desiredList = [line.rstrip() for line in infile]    
    desiredList = [line for line in infile]
    infile.close()
    return desiredList

def createNewFile(list_vice, list_pres, fileName):    
    outfile = open("AIP_05_Data Processing and File Access_data/"+fileName, 'w')
    set_vice = set(list_vice)
    set_pres = set(list_pres)
    set_both = set_vice.intersection(set_pres)
    outfile.writelines(set_both)
    outfile.close
    
main()
            
    
```
