```python
# def isAlpha(L):
# ## Determine whether list of lowercase words is in
# ## alphabetical order.
#     for i in range(len(L) - 1):
#         if L[i] > L[i + 1]:
#             return False
#         return True

def isAlpha(L):
    if len(L) <= 1: #종료조건 : 글자 줄이고 줄이고 비교할거 없을때까지 줄일때까지 알파벳 만족시 끝
        return True
    elif L[0] > L[1]: #첫번째가 두번째보다 크면 안되니까
        return False
    else:
        return isAlpha(L[1:]) #한칸씩 뒤로? 개수를 다음거로, 줄이는 방향으로
    
list_words1 = ['apple', 'banana', 'carrot']
list_words2 = ['zebra', 'zero', 'alpha']

print(isAlpha(list_words1))
print(isAlpha(list_words2))
```

```python
#교수님

def isAlphaRecursion(L):
    if len(L) <= 1:
        return True
    elif L[0] <= L[1]:
        return isAlphaRecursion(L[1:])
    else:
        return False

list_words1 = ['apple', 'banana', 'carrot']
list_words2 = ['zebra', 'zero', 'alpha']

print(isAlphaRecursion(list_words1))
print(isAlphaRecursion(list_words2))
```

```python
class Rectangle:
    def __init__(self, width=1, height=1):
        self._width = width
        self._height = height
        self.__secret = "나는 아이언맨"

    def setWidth(self, width):
        self._width = width

    def getWidth(self):
        return self._width

    def setHeight(self, height):
        self._height = height

    def getHeight(self):
        return self._height

    def area(self):
        return self._width * self._height

    def perimeter(self):
        return 2 * (self._width + self._height)

    def __str__(self):
        return ("Width: " + str(self._width) + "\nHeight: " + str(self._height)) 
    
rect = Rectangle(5,10)
print(rect) #투스트링 한 형태로 나온다.

#rect._width = 100 #이런식으로 하지 말고
rect.setWidth(100) #이렇게 하자. #이것도 막고싶다면?
#print(rect.__secret) #언더바 두줄 하니까 private 처리된다. 흉내만.
print(rect._Rectangle__secret) #직접 클래스 언급해주면 또 된다.
```
