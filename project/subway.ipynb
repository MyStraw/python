```python
df_drivinginfo = pd.read_csv('C:\coding\python\project\SubWayfile\변환\drivinginfo.csv')
df_stationinfo = pd.read_csv('C:\coding\python\project\SubWayfile\변환\stationinfo.csv')
df_platforminfo = pd.read_csv('C:\coding\python\project\SubWayfile\변환\platforminfo.csv')
df_facility = pd.read_csv("C:/coding/python/project/SubWayfile/변환/facility.csv")
df_count = pd.read_csv('C:\coding\python\project\SubWayfile\변환\count.csv')
```

```python
# Split the stations, arrival times, and departure times
stations_split = df_drivinginfo['운행구간정거장'].str.split('+', expand=True)
arrival_times_split = df_drivinginfo['정거장도착시각'].str.split('+', expand=True)
departure_times_split = df_drivinginfo['정거장출발시각'].str.split('+', expand=True)

# Initialize a new dataframe
new_data = pd.DataFrame()

# Repeat for each column (station)
for i in range(stations_split.shape[1]):
    temp_data = df_drivinginfo[['열차번호', '노선번호', '노선명', '운행구간기점명', '운행구간종점명', '운행유형', '요일구분', '운행속도', '운영기관전화번호', '데이터기준일자']].copy()
    temp_data['정거장'] = stations_split[i]
    temp_data['도착시각'] = arrival_times_split[i]
    temp_data['출발시각'] = departure_times_split[i]
    
    # Remove rows where station is None
    temp_data = temp_data.dropna(subset=['정거장'])
    
    # Append to the new dataframe
    new_data = pd.concat([new_data, temp_data])

# Reset the index of the new dataframe
new_data = new_data.reset_index(drop=True)

new_data.head()
```

```python
# Further split the station, arrival time, and departure time columns
new_data['정거장번호'], new_data['정거장명'] = new_data['정거장'].str.split('-', 1).str
new_data['도착시각번호'], new_data['도착시각'] = new_data['도착시각'].str.split('-', 1).str
new_data['출발시각번호'], new_data['출발시각'] = new_data['출발시각'].str.split('-', 1).str

# Remove the original columns
new_data = new_data.drop(columns=['정거장', '도착시각번호', '출발시각번호'])

new_data.head()
```
