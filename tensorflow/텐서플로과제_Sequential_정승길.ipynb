```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data(
    path='boston_housing.npz', test_split=0.2, seed=113
)

print(x_train.shape) #열이 몇개

model = tf.keras.Sequential([])
model.add(tf.keras.layers.Dense(units=1, input_dim=13))

model.summary()
model.compile(loss='mse', optimizer='adam')
record = model.fit(x_train, y_train, epochs=100, batch_size=10, verbose=2)


loss = record.history['loss']
print("loss:", loss[-1])
print("weights:", model.layers[0].weights[0].numpy())
print("bias:", model.layers[0].weights[1].numpy())

plt.plot(loss)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```
